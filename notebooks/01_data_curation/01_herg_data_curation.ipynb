# 01 - hERG Data Curation from ChEMBL

**Target:** CHEMBL240 (hERG potassium channel)

**Expected:** 8,000-10,000 curated compounds with pChEMBL values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from rdkit import Chem
from rdkit.Chem import Descriptors, Draw
from chembl_webresource_client.new_client import new_client
import sys
sys.path.insert(0, str(Path.cwd().parent.parent))
print('All imports OK')
```

## Step 1: Fetch hERG Bioactivity Data

```python
activity = new_client.activity
results = activity.filter(
    target_chembl_id='CHEMBL240',
    standard_type__in=['IC50', 'Ki', 'Kd'],
    pchembl_value__isnull=False
)
df = pd.DataFrame.from_records(results)
print(f'Total records fetched: {len(df)}')
df.head()
```

## Step 2: Data Exploration

```python
print(f'Unique compounds: {df["molecule_chembl_id"].nunique()}')
print(f'Activity types: {df["standard_type"].value_counts().to_dict()}')
fig, ax = plt.subplots(figsize=(10, 5))
df['pchembl_value'].astype(float).hist(bins=50, ax=ax, edgecolor='black')
ax.set_xlabel('pChEMBL Value')
ax.set_ylabel('Count')
ax.set_title('Distribution of hERG pChEMBL Values')
plt.tight_layout()
plt.show()
```
